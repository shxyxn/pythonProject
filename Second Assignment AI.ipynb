```python
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
# dataset :
x1 = np.random.rand(200,1) *8-4
x2= np.random.rand(200,1) *8-4
x3=np.random.rand(200,1) *8-4
x4=np.random.rand(200,1) *8-4
y1 =  (x1 **4 + x2**3 + x3**2 - x4 + 5) / 30
y1 = y1 + np.random.normal(scale=0.1, size=y1.shape)
plt.plot(x1, y1, 'bo')
```

```python
# split data
from sklearn.model_selection import train_test_split

# features = [x1,x2,x3,x4]
features = x1
label = y1
```

```python
features
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.2,random_state=101)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x_train,y_train, color='blue', label='Training set')
plt.scatter(x_test,y_test, color='c', label='Test set')
plt.title("THE DATA")
plt.legend(loc='best')
```

```python
# DEGREE = 1

poly_features = PolynomialFeatures(degree=1, include_bias=False)
x1_poly = poly_features.fit_transform(x_train)
reg1_1 = LinearRegression()
reg1_1.fit(x1_poly, y_train)
```

```python
# x1_vals = np.linspace(-4,4,5).reshape(-1,1)
x1_vals = x_train
x1_vals_poly = poly_features.transform(x1_vals)
y1_vals = reg1_1.predict(x1_vals_poly)
plt.suptitle('Degree 1')
plt.ylabel('Output')
plt.xlabel('Input')
plt.scatter(x_test, y_test, color="c")
plt.plot(x1_vals, y1_vals, color="r")
plt.show()
```

```python
print(f" MSE : {metrics.mean_squared_error(y_train, y1_vals)} \n R2 Score : {r2_score(y_train, y1_vals)}")
```

```python
from sklearn.model_selection import validation_curve
from sklearn.pipeline import make_pipeline
degrees = np.arange(1,21)

model = make_pipeline(PolynomialFeatures(), LinearRegression())

train_scores, validation_scores = validation_curve(
            model, x1, y1,
            param_name='polynomialfeatures__degree',
            param_range=degrees)

plt.figure(figsize=(6,4))
plt.plot(degrees, validation_scores.mean(axis=1), lw=2, label='cross-validation')
plt.plot(degrees, train_scores.mean(axis=1), lw=2, label='training')
plt.legend(loc='best')
plt.xlabel('degree of fiting')
plt.ylabel('Variance')
plt.title('Validation Curve')
plt.tight_layout()
```

```python
for k,v in a.items():

    
    model = make_pipeline(PolynomialFeatures(degree=k), LinearRegression())
    from sklearn.model_selection import learning_curve
    train_sizes, train_scores, validation_scores = learning_curve(
        model, x1 , y1,
        train_sizes=np.logspace(-1,0,20))

    plt.figure(figsize=(6,4))
    plt.plot(train_sizes, validation_scores.mean(axis=1),lw=2,color='c', label = 'cross-validation')
    plt.plot(train_sizes, train_scores.mean(axis=1),lw=2,color='b', label = 'Training')
    xj = 0,160
    yj= v,v
    plt.plot(xj, yj, lw=1, color='r',ls='--', label="R2 Score")
    plt.ylim(ymin=-.1, ymax=1)
    plt.legend(loc='best')
    plt.xlabel('number of train samples')
    plt.ylabel('Variance')
    plt.title(f'Learning Curve for degree: {k}')
    plt.tight_layout()
    

plt.show()
```
